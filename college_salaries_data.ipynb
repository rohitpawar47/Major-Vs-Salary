```python
sort_df = df.sort_values(by = 'Starting Median Salary', ascending = False)
sort_df = sort_df.reset_index()
sort_df.head()
```

```python
sort_df.drop('index',axis=1, inplace=True)
```

```python
sort_df.head()
```

```python
plt.figure(figsize=(10,5),dpi=150)
plt.barh(sort_df['Undergraduate Major'][:10], sort_df['Starting Median Salary'][:10], color='orange')
plt.xticks(fontsize=10, horizontalalignment='right')
plt.yticks(fontsize=10)
plt.ylabel('Undergraduate Major', fontsize=14, color='b')
plt.xlabel('Starting Median Salary', fontsize=14, color='b')
plt.title('Major Vs Starting Median Salary (Top 10 Result)', color='g')
plt.tight_layout()
```

* According to the insights above, **physician assistants** earn the **most money**, followed by **chemical engineers, computer engineers**, and so on.
* And the lowest erning degree is **spanish** followed by **religion, education** and so on.

# Finding highest and lowest erning degrees (Mid-Career)

```python
print(f"Highest Salary: {df['Mid-Career Median Salary'].max()}",","
      ,f"Index for the Highest Mid Career Salary: {df['Mid-Career Median Salary'].idxmax()}")

print(f"Lowest Salary: {df['Mid-Career Median Salary'].min()}",","
      ,f"Index for the Lowest Mid Career Salary: {df['Mid-Career Median Salary'].idxmin()}")
```

```python
df[['Undergraduate Major', 'Mid-Career Median Salary']].loc[[18]]
```

```python
groupby_mid_career = df.groupby('Undergraduate Major').agg({'Mid-Career Median Salary': pd.Series.sum})
```

```python
groupby_mid_career.nlargest(10,'Mid-Career Median Salary')
```

```python
large_mid_career = groupby_mid_career.nlargest(10,'Mid-Career Median Salary')
plt.figure(figsize=(10,5),dpi=150)
plt.barh(large_mid_career.index, large_mid_career['Mid-Career Median Salary'], color='orange')
plt.xticks(fontsize=10,horizontalalignment='right')
plt.yticks(fontsize=10)
plt.ylabel('Undergraduate Major', fontsize=14, color='b')
plt.xlabel('Mid-Career Median Salary', fontsize=14, color='b')
plt.title('Major Vs Mid-Career Median Salary (Top 10 Result)', color='g')
plt.tight_layout()
plt.show()
```

* According to the findings of the above investigation, **chemical engineering** has the **highest salary** after graduation, and **education** has the **lowest salary** after graduation.

# Finding The Lowest Risk Major

#### Lets find the difference

```python
df['Difference'] = df['Mid-Career 90th Percentile Salary'] - df['Mid-Career 10th Percentile Salary']
```

```python
df.head()
```

```python
diff = df[['Undergraduate Major','Starting Median Salary','Difference']].sort_values(by='Difference',ascending=True).head()
diff
```

```python
plt.figure(figsize=(10,5),dpi=150)
plt.barh(diff['Undergraduate Major'], diff['Difference'], color='maroon')
plt.xticks(fontsize=10,horizontalalignment='right')
plt.yticks(fontsize=10)
plt.ylabel('Undergraduate Major', fontsize=14, color='b')
plt.xlabel('Difference', fontsize=14, color='b')
plt.title('Major Vs Difference (Top 5 Result)', color='g')
plt.tight_layout()
plt.show()
```

* Based on the findings above, we may conclude that some **healthcare professions** graduates are more **confident about their wage prospects after graduation**.

# Finding Degrees with the highest potential

```python
high_potential = df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
high_potential[['Undergraduate Major','Mid-Career 90th Percentile Salary']].head()
```

```python
high_difference = df.sort_values('Difference', ascending=False)
high_difference[['Undergraduate Major','Difference']].head()
```

```python
df[['Undergraduate Major','Mid-Career Median Salary']].sort_values('Mid-Career Median Salary', ascending=False).head()
```

```python
percentile = high_potential[['Undergraduate Major','Mid-Career 90th Percentile Salary']].head(5)
plt.figure(figsize=(10,5),dpi=150)
plt.barh(percentile['Undergraduate Major'], percentile['Mid-Career 90th Percentile Salary'], color='maroon')
plt.xticks(fontsize=10,horizontalalignment='right')
plt.yticks(fontsize=10)
plt.ylabel('Undergraduate Major', fontsize=14, color='b')
plt.xlabel('Mid-Career 90th Percentile Salary', fontsize=14, color='b')
plt.title('Major Vs Mid-Career 90th Percentile Salary (Top 5 Result)', color='g')
plt.tight_layout()
plt.show()
```

* As we have discovered, the most promising **degrees** are **Economics, Finance, and Math** 

# Salaries by Group

```python
df.groupby('Group').count()
```

```python
pd.options.display.float_format = '{:,.2f}'.format
group_mean=df.groupby('Group').mean()
group_mean
```

```python
plt.figure(figsize=(10,5),dpi=150)
plt.barh(group_mean.index, group_mean['Mid-Career 90th Percentile Salary'], color='maroon')
plt.xticks(fontsize=10,horizontalalignment='right')
plt.yticks(fontsize=10)
plt.ylabel('Group', fontsize=14, color='b')
plt.xlabel('Mean of Mid-Career 90th Percentile Salary', fontsize=14, color='b')
plt.title('Group Vs Mean of Mid-Career 90th Percentile Salary', color='g', fontweight='bold')
plt.tight_layout()
plt.show()
```

* According to the abovementioned findings, **STEM has the greatest average salary**, followed by **Business**, while **HASS** has the lowest.
